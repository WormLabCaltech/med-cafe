```python
position = {}
for i, letter in enumerate(dpy):
    if letter in position.keys():
        position[letter] += [i]
    else:
        position[letter] = [i]

# triplet analysis
triplet = {}
nTriplets = 0
totalTriplets = 0
for i in range(len(dpy)-3):    
    trip = dpy[i: i + 3]
    if trip in triplet.keys():
        triplet[trip] += [i]
    else:
        triplet[trip] = [i]
        nTriplets += 1
    totalTriplets += 1
```

```python
maxT = ''
maxN = 0

for key, value in triplet.items():
    if len(value) > maxN:
        maxT = key
        maxN = len(value)
print(maxT)
```

```python
for letter, place in position.items():
    print(letter, len(place))
```

```python
array = []

for key in sorted(triplet):
    P1 = len(position[key[0]])
    P2 = len(position[key[1]])
    P3 = len(position[key[2]])
    probability = P1*P2*P3/len(dpy)**3
    expected = probability*totalTriplets
    array += [[key, len(triplet[key]), expected]]
triplet_df = pd.DataFrame(array, columns=['Triplet', 'Occurrence', 'Expected'])
triplet_df.sort_values('Occurrence', inplace=True)
```

```python
triplet_df['Surprise'] = (triplet_df.Occurrence - triplet_df.Expected)
```

```python
triplet_df.describe()
```

```python
sns.barplot(x='Occurrence', y='Triplet', data=triplet_df[triplet_df.Surprise < -0.85])
plt.title('Under-represented triplets')
```

```python
sns.barplot(x='Occurrence', y='Triplet', data=triplet_df[triplet_df.Surprise > 6])
plt.title('Over-represented triplets')
```
