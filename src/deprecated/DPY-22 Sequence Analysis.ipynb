```python
# doublet analysis
def polyplet(n):
    polyplet = {}
    nPolyplet = 0
    totalPolyplets = 0
    for i in range(len(dpy) - n):
        poly = dpy[i: i + n]
        rev_poly = dpy[i+n-1:i-1:-1]
        a =  np.array([i])
        if poly in polyplet.keys():
            polyplet[poly] = np.append(polyplet[poly], a)
#         if rev_poly in polyplet.keys():
#             polyplet[rev_poly] = np.append(polyplet[rev_poly], a)
        else:
            polyplet[poly] = a
            nPolyplet += 1
        totalPolyplets += 1

    return polyplet, nPolyplet, totalPolyplets

def n_random(key, position, total_polyplets, L=len(dpy)-1):
    P = np.array([len(position[k]) for k in key])
    probability = np.prod(P)/L**len(key)

    return probability*total_polyplets

def surprise(polyplet, position, total_polyplets):
    array = []
    for key in sorted(polyplet):
        expected = n_random(key, position, total_polyplets)
        array += [[key, len(polyplet[key]), expected]]

    df = pd.DataFrame(array, columns=['n-plet', 'Occurrence', 'Expected'])
    df['Surprise'] = (df.Occurrence - df.Expected)**2/df.Expected
    df['logSurprise'] = df.Surprise.apply(np.log10)
    df.sort_values('Surprise', inplace=True)
    
    return df
```

```python
singlet, nSinglet, totalSinglets = polyplet(1)
doublet, nDoublet, totalDoublets = polyplet(2)
triplet, nTriplet, totalTriplets = polyplet(3)
quadplet, nQuad, totalQuad = polyplet(4)
```

```python
singlet_df = surprise(singlet, singlet, totalSinglets)
doublet_df = surprise(doublet, singlet, totalDoublets)
triplet_df = surprise(triplet, singlet, totalTriplets)
quad_df = surprise(quadplet, singlet, totalQuad)
```
